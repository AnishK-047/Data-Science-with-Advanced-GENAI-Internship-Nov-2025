## 2. Python If-Else

```python
n = int(input().strip())


if n % 2 != 0:
    print("Weird")
elif n % 2 == 0 and n in range(2,5+1):
    print("Not Weird")
elif n % 2 ==0 and n in range (6,20+1):
    print("Weird")
else:
    print("Not Weird")
```

## 6. Write a function

```python
def is_leap(year):
    leap = False
    
    # Write your logic here
    if year % 400 == 0:
        return "True"
    if year % 100 == 0:
        return "False"
    if year % 4 == 0:
        return "True"
    
    return leap

year = int(input())
print(is_leap(year))
```
