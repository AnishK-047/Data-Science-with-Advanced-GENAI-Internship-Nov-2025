## 1. Given the array of integers nums, you will choose two different indices i and j of that array. Return the maximum value of (nums[i]-1)*(nums[j]-1).

```python
class Solution:
    def maxProduct(self, nums: list[int]) -> int:
        max1 = max2 = 0

        for num in nums:
            if num > max1:
                max2 = max1
                max1 = num
            elif num > max2:
                max2 = num

        return (max1 - 1) * (max2 - 1)

a= Solution()
print(a.maxProduct([3,4,5,2]))
```

## 2. There are n soldiers standing in a line. Each soldier is assigned a unique rating value.

## You have to form a team of 3 soldiers amongst them under the following rules:

## Choose 3 soldiers with index (i, j, k) with rating (rating[i], rating[j], rating[k]).
## A team is valid if: (rating[i] < rating[j] < rating[k]) or (rating[i] > rating[j] > rating[k]) where (0 <= i < j < k < n).

```python
class Solution:
    def numTeams(self, rating: list[int]) -> int:
        n = len(rating)
        teams = 0

        for j in range(n):
            left_smaller = left_greater = 0
            right_smaller = right_greater = 0

            # check left side of j
            for i in range(j):
                if rating[i] < rating[j]:
                    left_smaller += 1
                else:
                    left_greater += 1

            # check right side of j
            for k in range(j + 1, n):
                if rating[k] > rating[j]:
                    right_greater += 1
                else:
                    right_smaller += 1

            # increasing + decreasing teams
            teams += left_smaller * right_greater
            teams += left_greater * right_smaller

        return teams
b= Solution()
print(b.numTeams([2,5,3,4,1]))
```

## 3. Given two integer arrays startTime and endTime and given an integer queryTime. The ith student started doing their homework at the time startTime[i] and finished it at time endTime[i].

```python
class Solution:
    def busyStudent(self, startTime: list[int], endTime: list[int], queryTime: int) -> int:
        count = 0

        for i in range(len(startTime)):
            if startTime[i] <= queryTime <= endTime[i]:
                count += 1

        return count
c= Solution()
print(c.busyStudent([1,2,3], endTime = [3,2,7], queryTime = 4))
```

## 4. Given an integer num, return the number of steps to reduce it to zero.In one step, if the current number is even, you have to divide it by 2, otherwise, you have to subtract 1 from it.

```python
class Solution:
    def numberOfSteps(self, num: int) -> int:
        steps = 0

        while num > 0:
            if num % 2 == 0:
                num //= 2
            else:
                num -= 1
            steps += 1

        return steps

d= Solution()
print(d.numberOfSteps(14))
```

## 5. Given an integer n, return an array ans of length n + 1 such that for each i (0 <= i <= n), ans[i] is the number of 1's in the binary representation of i.

```python
class Solution:
    def countBits(self, n: int) -> list[int]:
        ans = [0] * (n + 1)

        for i in range(1, n + 1):
            ans[i] = ans[i // 2] + (i % 2)

        return ans

E= Solution()
print(E.countBits(2))
```
